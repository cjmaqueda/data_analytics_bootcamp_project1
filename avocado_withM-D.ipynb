```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
%pylab inline
```

```python
avocado = pd.read_csv('avocado_data_MarketStateRegion_plus one day.csv',parse_dates = True, index_col = 'Date')
del avocado['Unnamed: 0']
avocado_data = avocado.rename(columns={"4046":"small volume sold","4225":"large volume sold","4770":"extra large volume sold"})
avocado_data.head()
```

```python
avocado_data['year'] = None
avocado_data['month'] = None

for index, row in avocado_data.iterrows():
    avocado_data.at[index, 'year'] = str(index.year)
    avocado_data.at[index, 'month'] = str(index.month)
```
